# SpectraManager Plotting Demo

This notebook demonstrates how to use the `SpectraManager` along with the
plotting utilities in `sbi_delta.spectra.plotting`.


```python
import os
import sys
base_path = os.path.abspath("../..")
print(f"Adding base path: {base_path}")
sys.path.append(base_path)  # Adjust path as needed
from sbi_delta.config import BaseConfig
from sbi_delta.spectra_manager import SpectraManager
```

```python
# Initialize with your data folder and dye names
spectra = os.path.join(base_path, 'data', 'spectra_npz')
cfg = BaseConfig(min_wavelength=300.0, max_wavelength=700.0, wavelength_step=1.0, 
                 spectra_folder=spectra, dye_names=['AF488', 'AF647'], bg_dye='AF_v1')
mgr = SpectraManager(cfg)
mgr.load()
print(mgr)
```

```python
# Plot a single emission spectrum
ax = mgr.plot_emission('AF488')
ax.figure.show()
```

```python
# Plot all emission spectra
ax =  mgr.plot_all_emissions()
ax.figure.show()
```

```python
# Plot a single excitation spectrum
ax = mgr.plot_excitation('AF647')
ax.figure.show()
```

```python
# Plot all excitation spectra
ax = mgr.plot_all_excitations()
ax.figure.show()
```

```python
ax = mgr.plot_all_spectra();
```
