# Visualizing the Joint Prior in SBI-DELTA

This notebook demonstrates how to use the PriorManager to visualize the joint prior (concentrations + background ratio) for a given configuration.

```python
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
base_path = os.path.abspath("../..")
sys.path.append(base_path)  # Adjust path as needed
from sbi_delta.config import BaseConfig, PriorConfig
from sbi_delta.prior_manager import PriorManager
```

```python
# Example configuration
spectra_folder = os.path.join(base_path, "data/spectra_npz")  # adjust path as needed
base_config = BaseConfig(dye_names=['JF479','JF552', 'JF646'], bg_dye='AF_v1', spectra_folder=spectra_folder)
prior_config = PriorConfig(dirichlet_concentration=5.0, include_background_ratio=True, background_ratio_bounds=(0.1, 0.2))
prior_manager = PriorManager(prior_config, base_config)
print(prior_manager)
```

```python
# Visualize joint prior (concentrations and background ratio)
axes = prior_manager.visualize_joint_prior(n_samples=20000)
```

```python
# Example configuration 2
base_config = BaseConfig(dye_names=['JF479','JF552', 'JF646', 'JFX650'], bg_dye='AF_v1', spectra_folder=spectra_folder)
prior_config = PriorConfig(dirichlet_concentration=3, include_background_ratio=True, background_ratio_bounds=(0.2, 0.4))
prior_manager = PriorManager(prior_config, base_config)
print(prior_manager)
axes = prior_manager.visualize_joint_prior(n_samples=20000)
```
