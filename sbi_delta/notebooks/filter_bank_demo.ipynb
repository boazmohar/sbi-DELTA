# FilterBank Demonstration

This notebook shows how to configure and use the `FilterBank` class
from `sbi_delta.filter_bank`, including repr, single-filter plotting, and
overlaying all filter responses.


```python
import sys
import os
sys.path.append(os.path.abspath("../.."))  # Adjust path as needed
import numpy as np
import matplotlib.pyplot as plt
from sbi_delta.config import FilterConfig
from sbi_delta.filter_bank import FilterBank
```

```python
from sbi_delta.config import BaseConfig, FilterConfig
from sbi_delta.filter_bank import FilterBank

# 1) Define the wavelength grid via BaseConfig
base_cfg = BaseConfig(min_wavelength=400, max_wavelength=700, wavelength_step=1)

# 2) Create filter configs
f1 = FilterConfig(start=420, stop=450, sharpness=1)
f2 = FilterConfig(start=480, stop=510, sharpness=2)

# 3) Build the bank
bank = FilterBank(base_cfg, [f1, f2])
print(bank)
```

```python

# 4) Plot individual or all filters
ax1 = bank.plot_filter(0)
ax2 = bank.plot_all_filters()
plt.show()
```

```python

# 5) Change filters on the fly
f3 = FilterConfig(start=520, stop=550, sharpness=4)
bank.set_filters([f1, f3])
print(bank)
ax3 = bank.plot_all_filters()
```
